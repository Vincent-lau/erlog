```python
import networkx as nx
import matplotlib.pyplot as plt
import os
```

```python
SEED = 9 # enter your seed
constructor = [(5, 10, 0.5), (5, 10, 0.1)]
G = nx.random_shell_graph(constructor, seed = SEED)
fig, ax = plt.subplots()
fig.set_size_inches((16,9))
nx.draw(G, ax = ax, with_labels = True)

```

```python
with open(f'../bench_program/tc_program.dl', 'r') as in_file, \
  open(f'../bench_program/tc_bench.dl', 'w') as out_file:

  for line in in_file:
    out_file.write(line)

  out_file.write('\n')

  for (x, y) in G.edges:
    out_file.write(f'link("a{x}", "a{y}").\n')
```
